````python
import cv2
````

# Otworzenie zdjęć

````python
sciezka_do_zdjecia = "Roza.jpg"
zdjecie = cv2.imread(sciezka_do_zdjecia)
````

Wyświetlenie zdjęcia

````python
import matplotlib.pyplot as plt



zdjecie_rgb = cv2.cvtColor(zdjecie, cv2.COLOR_BGR2RGB)#zrzutowanie z BGR na RGB
    
plt.figure()
plt.imshow(zdjecie_rgb)
plt.title("Moje zdjęcie")
plt.axis('off')
plt.show()
````

# Upscaling

````python
nowe_wymiary = (80,80)
zdjecie_resize = cv2.resize(zdjecie, nowe_wymiary)
zdjecie_resize_rgb = cv2.cvtColor(zdjecie_resize, cv2.COLOR_BGR2RGB)#zrzutowanie z BGR na RGB
    
plt.figure()
plt.imshow(zdjecie_resize_rgb)
plt.title("Moje zdjęcie")
plt.axis('off')
plt.show()
````

Upscaling wykonuje się podobnie jak `resize`, ale trzeba podać kilka dodatkowych danych:

```python
cv2.resize(zdjecie, None, fx=powiększenie_x, fy=powiększenie_y, interpolation=rodzaj_interpolacji)
```

powiększenie_x, powiększenie_y – określają, ile razy zwiększyć ilość pikseli w poziomie i pionie.  
rodzaj_interpolacji – określa rodzaj interpolacji, do wyboru mamy:

    - cv2.INTER_NEAREST – interpolacja najbliższego sąsiada (szybka, ale niska jakość)
    - cv2.INTER_LINEAR   – interpolacja biliniowa (domyślna przy powiększaniu)
    - cv2.INTER_CUBIC    – interpolacja bikubiczna (wolniejsza, lepsza jakość)
    - cv2.INTER_LANCZOS4 – interpolacja Lanczos z 8 sąsiadami (najlepsza jakość, najwolniejsza)

````python
nowe_wymiary = (800,800)
zdjecie_resize = cv2.resize(zdjecie_resize, None, fx=2.0, fy=2.0, interpolation=cv2.INTER_LINEAR)
zdjecie_resize_rgb = cv2.cvtColor(zdjecie_resize, cv2.COLOR_BGR2RGB)#zrzutowanie z BGR na RGB
    
plt.figure()
plt.imshow(zdjecie_resize_rgb)
plt.title("Moje zdjęcie")
plt.axis('off')
plt.show()
````
